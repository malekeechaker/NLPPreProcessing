```python
#2nd methode code from scratch
import pandas as pd
data=[["Revue 1" , "Spider man est un film incroyable !"],
["Revue 2" , "Spider man est le meilleur film !"],
["Revue 3" , "Spiderman est TELLEMENT génial."]]
df=pd.DataFrame (data,columns=["revue", "text"])
df
```

```python
from keras.preprocessing.text import text_to_word_sequence

def tokenise_text(text):

  return text_to_word_sequence(text)

#convertir la phrase en miniscule
#supprimer la ponctuation
#tokanization par espace
df["text_tok"]=df["text"].apply(tokenise_text)
df
```

```python

def calculBOW(document):
    bow = {}
    mots = document.split()
    for mot in mots:
        if mot in bow:
            bow[mot] += 1
        else:
            bow[mot] = 1
    return bow

df["text_bow"] = df["text"].apply(calculBOW)
df
```

```python
#3rd methode countvectorizer
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vecteurs_bow = vectorizer.fit_transform(df["text"])

#les mots du vocabulaire
print( vectorizer.get_feature_names_out())

#les vecteurs BOW
print( vecteurs_bow.toarray())

#dir(vecrorizer)  methodes
```

```python
#from scrach and delete stop words
stop_words = ['le', 'un', 'est', 'le', 'un']

def calculBOW(document, stop_words):
    bow = {}
    mots = document.split()
    for mot in mots:
        if mot not in stop_words:
            if mot in bow:
                bow[mot] += 1
            else:
                bow[mot] = 1
    return bow
df["text_bow_nostop"] = df["text"].apply(lambda x: calculBOW(x, stop_words))
df
```

```python
#4th methode Bag of bigrams

vectorizer_bigrams = CountVectorizer(ngram_range=(1, 2)) #creer signle and bigram ngram_range=(start,finish)
vecteurs_bigrams = vectorizer_bigrams.fit_transform(df["text_tok"].apply(' '.join))

print( vectorizer_bigrams.get_feature_names_out())

print( vecteurs_bigrams.toarray())
```

```python
import re

#read a text from file
with open('alice.txt', 'r') as file:
    content = file.readlines()
    new_content = [cont.replace('\n'," ") for cont in content]

print(new_content)
```

```python
from keras.preprocessing.text import text_to_word_sequence

#tokanization
tokenized_phrases = [text_to_word_sequence(phrase) for phrase in new_content]
tokenized_phrases
```

```python
#5th methode TF-IDF (term frequency-inverse document frequency) = TF * IDF
#IDF = log( nbr doc/nbr doc with term )
#TF in doc = nbr term t in doc / nbr of termss in doc

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_TDIDF = TfidfVectorizer()
vectorizer_TDIDF.fit(new_content)
transform_output = vectorizer_TDIDF.transform(new_content)

#les mots du vocabulaire
print( vectorizer_TDIDF.get_feature_names_out())

#les vecteurs TF-IDF
print( transform_output.toarray())
```

# **Word Embeding**

capture similarity in the vectors

lower dimensions than BOW

```python
#1st methode Word2Vec CBOW
from gensim.models import Word2Vec

#cbow (continous bag of words) use neural natwork with the context is the input and the word is the output (to determine the vector of a word which is the weights)
model_cbow = Word2Vec(tokenized_phrases, vector_size=100, window=5, min_count=2, sg=0)

print("Vecteur du mot 'alice':", model_cbow.wv['alice'])
print("3 mots les plus similaires à 'alice':", model_cbow.wv.most_similar('alice', topn=3))

```

```python
#2nd methode Word2Vec SkipGram
#skipgram also use neural natwork, but with the word is the input and the context is the output (to determine the vector of a word which is the weights)
model_skipgram = Word2Vec(tokenized_phrases, vector_size=100, window=5, min_count=2, sg=1)

print("Vecteur du mot 'alice':", model_skipgram.wv['alice'])
print("3 mots les plus similaires à 'alice':", model_skipgram.wv.most_similar('alice', topn=3))
```

```python
import numpy as np

# Define the words for which you want to calculate cosine similarity
words = ['wonderland', 'alice', 'rabbit']

# Calculate and print cosine similarity for CBOW model
print("Cosine Similarity for CBOW Model:")
for i in range(len(words)):
    for j in range(i+1, len(words)):
        similarity = model_cbow.wv.similarity(words[i], words[j])
        print(f"Similarity between '{words[i]}' and '{words[j]}': {similarity}")

```

```python
# Calculate and print cosine similarity for skipgram model
print("Cosine Similarity for skipgram Model:")
for i in range(len(words)):
    for j in range(i+1, len(words)):
        similarity = model_skipgram.wv.similarity(words[i], words[j])
        print(f"Similarity between '{words[i]}' and '{words[j]}': {similarity}")
```

```python

mot_cible = "wonderland"

# Trouver les mots les plus similaires avec CBOW
print(f"Les mots les plus similaires à {mot_cible} par CBOW :")
similaires_cbow = model_cbow.wv.most_similar(mot_cible, topn=5)
for word, similarity in similaires_cbow:
    print(f"{word}: {similarity}")

# Trouver les mots les plus similaires avec Skip-gram
print(f"Les mots les plus similaires à {mot_cible} par Skip-gram :")
similaires_skipgram = model_skipgram.wv.most_similar(mot_cible, topn=5)
for word, similarity in similaires_skipgram:
    print(f"{word}: {similarity}")
```

```python
#3rd methode Word2Vec spacy
import spacy

# word vectors you need to install large or medium english model. We will install the large one!
nlp = spacy.load("en_core_web_lg")
```

```python
def print_similarity(base_word, words_to_compare):
    base_token = nlp(base_word)
    doc = nlp(words_to_compare)
    for token in doc:
        print(f"{token.text} <-> {base_token.text}: ", token.similarity(base_token))

print_similarity("iphone", "apple samsung iphone dog kitten")
```

```python
king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
woman = nlp.vocab["woman"].vector
queen = nlp.vocab["queen"].vector

result = king - man + woman
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([result], [queen])
```
